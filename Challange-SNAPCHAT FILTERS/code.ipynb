```python
import cv2
import pandas as pd
import matplotlib.pyplot as plt
# input = cv2.imread("./Train/Jamie_Before.jpg")
input = cv2.imread("./Test/Before.png")
s_img = cv2.imread("./Train/glasses.png",-1)
s_img2 = cv2.imread("./Train/mustache.png",-1)

nose_cascade = cv2.CascadeClassifier("./Train/third-party/Nose18x15.xml")
eyes_cascade = cv2.CascadeClassifier("./Train/third-party/frontalEyes35x16.xml")

noses = nose_cascade.detectMultiScale(input,1.3,5)
eyes  = eyes_cascade.detectMultiScale(input,1.3,5)


for (x,y_,w,h) in noses:
    mesam = cv2.resize(s_img2,(w,h))
    y = y_+30
    for i in range(mesam.shape[0]):
        for j in range(mesam.shape[1]):
            if mesam[i,j,3] > 0:
                input[y+i,x+j,:] = mesam[i,j,:-1]

for (x,y,w,h) in eyes:
    s_img = cv2.resize(s_img,(w,h))
    for i in range(s_img.shape[0]):
        for j in range(s_img.shape[1]):
            if s_img[i,j,3] > 0:
                input[y+i,x+j,:] = s_img[i,j,:-1]

plt.imshow(input)
print(input.shape)
Person_x = input.reshape((-1,3))

df = pd.DataFrame(Person_x, columns=["Channel 1","Channel 2", "Channel 3"])

df.to_csv("pred1.csv", index=False)
```
